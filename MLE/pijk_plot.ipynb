```python
import numpy as np
import random
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
import math
from scipy.optimize import minimize, fsolve, leastsq
from scipy.special import gamma, erf, erfinv
from scipy.stats import pearsonr, poisson, f
from MLE_functions import *

warnings.filterwarnings("ignore")
```

```python
# Input parameters
sigma = 1.02
B = 350
kd_guesses = [50, 70, 100]
fmax_guesses = [40000, 30000, 20000]
conc_to_ignore = [0, 25, 2000]
# input_csv = "/projects/brpe7306/fab-library-barcoding/MLE/input/4A8_CC121_combined_ref.csv"
input_csv = "/projects/brpe7306/fab-library-barcoding/Match_all/_combined.csv"
output_csv = "/projects/brpe7306/fab-library-barcoding/MLE/output/test.csv"


input_df = pd.read_csv(input_csv)
# bcs = input_df['Barcode'].unique()[:40]
# input_df = input_df[input_df['Barcode'].isin(bcs)]
bcs = input_df['Barcode'].unique()[0:40]
input_df = input_df[input_df["Barcode"].isin(bcs)]

# Run MLE
start = time.time()
output = run_mle(
    input_df,
    conc_to_ignore,
    sigma=sigma,
    B=B,
    kd_guesses=kd_guesses,
    fmax_guesses=fmax_guesses,
    variant_col="Barcode",
    top25_only=True,
    highmid_only=False,
    plot=True,
    print_df=False,
    verbose=False,
    status=True,
)
end = time.time()
output.to_csv(output_csv, index=False)
print(output)
print("Finished in", round((end - start), 2), "seconds.")
```
