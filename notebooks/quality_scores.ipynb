```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
score_names = [["Q20", 20], ["Q15", 15], ["Q10", 10], ["Q5", 5]]


def convert_q_scores(q):
    q_scores = np.array(bytearray(q, "ascii"), dtype=np.int8)
    q_scores = [i - 33 for i in q_scores]
    return q_scores


def print_seq_by_quality(s, q, print_qual=False):
    if print_qual:
        for i in range(len(q)):
            if q[i] > score_names[0][1]:
                print(q[i], end=" ")
            elif q[i] > score_names[1][1]:
                print("\x1b[1;30;47m" + str(q[i]) + "\x1b[0m", end=" ")
            elif q[i] > score_names[2][1]:
                print("\x1b[6;30;42m" + str(q[i]) + "\x1b[0m", end=" ")
            elif q[i] > score_names[3][1]:
                print("\x1b[0;37;43m" + str(q[i]) + "\x1b[0m", end=" ")
            else:
                print("\x1b[0;37;41m" + str(q[i]) + "\x1b[0m", end=" ")

    for i in range(len(q)):
        if q[i] > score_names[0][1]:
            print(s[i], end="")
        elif q[i] > score_names[1][1]:
            print("\x1b[1;30;47m" + str(s[i]) + "\x1b[0m", end="")
        elif q[i] > score_names[2][1]:
            print("\x1b[6;30;42m" + str(s[i]) + "\x1b[0m", end="")
        elif q[i] > score_names[3][1]:
            print("\x1b[0;37;43m" + str(s[i]) + "\x1b[0m", end="")
        else:
            print("\x1b[0;37;41m" + str(s[i]) + "\x1b[0m", end="")
    print("")


def print_legend():
    print("Legend:", end=" ")
    print("\x1b[1;30;47m" + str(score_names[0][0]) + "\x1b[0m", end=" ")
    print("\x1b[6;30;42m" + str(score_names[1][0]) + "\x1b[0m", end=" ")
    print("\x1b[0;37;43m" + str(score_names[2][0]) + "\x1b[0m", end=" ")
    print("\x1b[0;37;41m" + str(score_names[3][0]) + "\x1b[0m", end=" ")


df = pd.read_csv("/projects/brpe7306/fab-library-barcoding/maps/VH2/merged.csv")
# df = df[df["Amplen"] == 412]
# df = df[df['Seq'].str.contains('TCAACG')]

print_legend()
for i in range(40):
    q = convert_q_scores(df["Qual"].iloc[i])
    s = df["Seq"].iloc[i]
    print_seq_by_quality(s, q)

# sns.histplot(q)
```
