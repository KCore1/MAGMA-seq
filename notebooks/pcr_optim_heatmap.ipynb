```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
variant = "4A8>L:S7T-ACT"
barcode = "CCGTCCTCATTTCCTTCTGT"
print("Variant:", variant, "Barcode:", barcode)
pcr_cond = {}
sample_map = {
    1: "Q5 min rSAP 3",
    2: "Control Q5 25 rSAP 1",
    4: "Phusion min column clean 3",
    5: "Control Q5 25 rSAP 3",
    6: "Phusion min column clean 2",
    7: "KAPA min column clean 1",
    8: "Phusion min rSAP 2",
    9: "Q5 min column clean 2",
    10: "KAPA min column clean 2",
    11: "IML Control Q5 25 rSAP 1",
    13: "Q5 min rSAP 1",
    14: "Q5 min column clean 3",
    15: "Phusion min rSAP 3",
    16: "IML Control Q5 25 rSAP 2",
    18: "Phusion min column clean 1",
    19: "Q5 min column clean 1",
    20: "KAPA min rSAP 1",
    22: "Control Q5 25 rSAP 2",
    23: "KAPA min rSAP 2",
    24: "Q5 min rSAP 2",
    25: "IML Control Q5 25 rSAP 3",
    26: "KAPA min rSAP 3",
    27: "KAPA min column clean 3",
    28: "Phusion min rSAP 1",
}

for i in range(28):
    try:
        df = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/4A8VL"
            + str(i + 1)
            + "/map.csv"
        )
        df_merge = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/4A8VL"
            + str(i + 1)
            + "/merged.csv"
        )
        # print("Sample MBK" + str(i+1))
        total = df["Count"].sum()
        gene = df[df["Mutations"] == variant]["Count"].sum()
        bc = df[(df["Barcode"] == barcode)]["Count"].sum()
        gene_bc = df[(df["Barcode"] == barcode) & (df["Mutations"] == variant)][
            "Count"
        ].sum()
        # print("Total paired:", total)
        # print("Total with correct gene:", gene/total)
        # print("Total paired to barcode:", bc/total)
        # print("Total paired to barcode with correct gene:", gene_bc/total)
        pair_pct = [gene / bc, gene_bc / bc]
        pcr_cond[sample_map[i + 1]] = pair_pct
    except:
        continue

df_heatmap = pd.DataFrame(
    pcr_cond, index=["Correct Gene (4A8:S7T)", "Correct Pairing (4A8:S7T-CCGT)"]
)
df_heatmap = df_heatmap.reindex(sorted(df_heatmap.columns), axis=1)

variant = "4A8>L:S7T-ACT;T94M-ATG;D110E-GAG"
barcode = "ACATATTCAAAGACGGCAGG"
print("Variant:", variant, "Barcode:", barcode)
for i in range(28):
    try:
        df = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/4A8VL"
            + str(i + 1)
            + "/map.csv"
        )
        df_merge = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/4A8VL"
            + str(i + 1)
            + "/merged.csv"
        )
        total = df["Count"].sum()
        gene = df[df["Mutations"] == variant]["Count"].sum()
        bc = df[(df["Barcode"] == barcode)]["Count"].sum()
        gene_bc = df[(df["Barcode"] == barcode) & (df["Mutations"] == variant)][
            "Count"
        ].sum()
        # print("Total paired:", total)
        # print("Total with correct gene:", gene/total)
        # print("Total paired to barcode:", bc/total)
        # print("Total paired to barcode with correct gene:", gene_bc/total)
        pair_pct = [gene / bc, gene_bc / bc]
        pcr_cond[sample_map[i + 1]] = pair_pct
    except:
        continue

df_heatmap2 = pd.DataFrame(
    pcr_cond,
    index=["Correct Gene (4A8:S7T,T94M)", "Correct Pairing (4A8:S7T,T94M-ACAT)"],
)
df_heatmap2 = df_heatmap2.reindex(sorted(df_heatmap2.columns), axis=1)

df_heatmap = df_heatmap.append(df_heatmap2)

variant = "CC121>L:WT"
barcode = "CTTGATGTCTGCGATTCTTG"
print("Variant:", variant, "Barcode:", barcode)
for i in range(28):
    try:
        df = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/CC121VL"
            + str(i + 1)
            + "/map.csv"
        )
        df_merge = pd.read_csv(
            "/projects/brpe7306/fab-library-barcoding/PCR_VL/CC121VL"
            + str(i + 1)
            + "/merged.csv"
        )
        total = df["Count"].sum()
        gene = df[df["Mutations"] == variant]["Count"].sum()
        bc = df[(df["Barcode"] == barcode)]["Count"].sum()
        gene_bc = df[(df["Barcode"] == barcode) & (df["Mutations"] == variant)][
            "Count"
        ].sum()
        # print("Total paired:", total)
        # print("Total with correct gene:", gene/total)
        # print("Total paired to barcode:", bc/total)
        # print("Total paired to barcode with correct gene:", gene_bc/total)
        pair_pct = [gene / bc, gene_bc / bc]
        pcr_cond[sample_map[i + 1]] = pair_pct
    except:
        continue

df_heatmap2 = pd.DataFrame(
    pcr_cond, index=["Correct Gene (CC121:WT)", "Correct Pairing (CC121:WT-CTTG)"]
)
df_heatmap2 = df_heatmap2.reindex(sorted(df_heatmap2.columns), axis=1)

df_heatmap = df_heatmap.append(df_heatmap2)

custom_params = {
    "axes.spines.right": False,
    "axes.spines.top": False,
    "figure.dpi": 150,
    "savefig.dpi": 300,
}
sns.set_theme(
    rc=custom_params, style="dark", palette=None, font="sans serif", font_scale=1.2
)
# plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 3))
cbar_kws = {"orientation": "vertical", "shrink": 1.1, "extendfrac": 0.1}
annot_kws = {"fontsize": 9, "weight": "bold"}
sns.heatmap(
    df_heatmap,
    annot=True,
    fmt=".2g",
    vmin=0.7,
    vmax=1,
    cmap="Blues",
    annot_kws=annot_kws,
    cbar=False,
    cbar_kws=cbar_kws,
)
plt.savefig("pcr_optim_heatmap.png", bbox_inches="tight")
plt.show()
```
